### Insights

#### I- What is the most common room type in NYC Airbnb listing?



```python
# Count room types
most_commom = room_types['room_type'].value_counts()
most_commom
```

#### II- What is the average price of a listing by room type? 

```python
#Joining tables prices and room_type (price_rt)
price_rt = pd.merge(room_types, prices, on='listing_id', how='outer')
pd.DataFrame(price_rt)
```

```python
#Calculate average price by room type
avg_price_rt = price_rt.groupby('room_type')['price'].mean().round(2)
pd.DataFrame(avg_price_rt)
```

#### III- Which borough has the highest average price per month?

```python
#Calculate the borough average price per month (borough_ppm)
borough_ppm = prices.groupby('borough')['price_per_month'].mean().round(2)
borough = borough_ppm.sort_values(ascending=False)
pd.DataFrame(borough)
```

```python
#Calculate the neighbourhood average price per month (neighbourhood_ppm)
neighbourhood_ppm = prices.groupby('neighbourhood')['price_per_month'].mean().round(2)
neighbourhood = neighbourhood_ppm.sort_values(ascending=False)
pd.DataFrame(neighbourhood)
```

#### IV- How many listings of each room type are in each borough? 

```python
# Amount of listings of each room by borough
borough_rt = price_rt.groupby(['borough', 'room_type'])['listing_id'].count()
pd.DataFrame(borough_rt)
```

#### V- How many listings in each room type category have a price of over $500 per night?

```python
# Filter for prices greater than $500
over_500 = price_rt[price_rt['price'] > 500]
over_500_rt = over_500.groupby(['room_type'])['room_type'].count() 
pd.DataFrame(over_500_rt)
```

#### VI- What is the distribution of listing prices by neighborhood?

```python

dist_price = price_rt.groupby(['borough', 'room_type'])['price'].agg(['min', 'max', 'mean']).round(2)
dist_price
```

#### VII- What is the estimated amount of revenue generated by hosts in each borough?

```python
# Joining tables price with reviews
price_rvw = pd.merge(reviews, prices, on='listing_id', how='outer')
pd.DataFrame(price_rvw)
```

```python
# Ensure there is at least 1 rent per year
booked_days = price_rvw[price_rvw['booked_days_365'] > 0]

# Borough revenue by host 
rev_host = booked_days.groupby(['borough'])['price'].sum()
pd.DataFrame(rev_host)
```

#### VIII- How many listings have no reviews?

```python
# Ensure no missing values on column
no_reviews = reviews['number_of_reviews'].isna().sum()

# Cheking if there is no review at all
no_reviews = reviews[reviews['number_of_reviews'] == 0].sum().sum()
no_reviews
```

#### IX- How do the estimated book days correlate with the price of an Airbnb listing in New York City?

```python
correlation = price_rvw['price'].corr(price_rvw['booked_days_365'])
print(correlation)
```

###### A weak correlation is identified once the value is: -0.07. Meaning that as the price increases, the number of booked days tends to decrease.


* Negative -1 

* Weak      0

#### X- How customer are related to their own units

```python
# Function to filter clients who have units booked 365 days (full_yb)
def fyb_filter(reviews, booked_days_365):
    filtered_fyb = reviews[reviews['booked_days_365'] == 365]
    return filtered_fyb
```

```python
# Full year booked units(fyb)
fyb = fyb_filter(reviews, 'booked_days_365')
fyb
```

```python
# Amount of the units full year booked.
fyb_all = fyb['listing_id'].count()
fyb_all

```

##### Analysis with customer groups based on the number of days the unit is rented per year.

```python
#  Amount of full year booked units by Borough
fyb_borough = fyb_filter(price_rvw, 'booked_days_365').groupby(['borough'])['listing_id'].count()
df_fyb_borough = pd.DataFrame({'Borough': fyb_borough.index, 'Units 365 days Booked': fyb_borough.values}).set_index('Borough', drop=True).sort_values('Units 365 days Booked', ascending=False)
pd.DataFrame(df_fyb_borough)
```

```python
# Amount of units full year booked(fyb) grouped by borough and neighbourhood
fyb_borough_neighbourhood = fyb_filter(price_rvw, 'booked_days_365').groupby(['borough', 'neighbourhood'])['listing_id'].count()
df_fyb_borough_neighbourhood = pd.DataFrame({'borough': fyb_borough_neighbourhood.index.get_level_values('borough'),
                                             'neighbourhood': fyb_borough_neighbourhood.index.get_level_values('neighbourhood'),
                                             '365 days Booked': fyb_borough_neighbourhood.values})
pd.DataFrame(df_fyb_borough_neighbourhood)
```

```python
# Amount collected monthly by borough per units booked 365 days, full year(fyb)
fyb_borough = fyb_filter(price_rvw, 'booked_days_365').groupby(['borough'])['price_per_month'].sum().round(2)
df_fyb_borough_ppm = pd.DataFrame({'borough': fyb_borough.index, 'Monthly collectd fyb units': fyb_borough.values}).set_index('borough', drop=True).sort_values('Monthly collectd fyb units', ascending=False)
pd.DataFrame(df_fyb_borough_ppm)
```
